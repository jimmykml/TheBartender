# 🧪 The Bartender Demo
This notebook demonstrates how to use the full agent pipeline to analyze a stock using short-term and long-term signals.

```python
%load_ext autoreload
%autoreload 2

# ✅ Step 1: Load environment variables
import os
from dotenv import load_dotenv

load_dotenv()
```

```python
# ✅ Step 2: Setup import path (assuming notebook is in /notebooks)
import sys
import os

project_root = os.path.abspath(os.path.join(os.getcwd(), ".."))
if project_root not in sys.path:
    sys.path.append(project_root)
```

```python
# ✅ Step 3: Import orchestrator and model
from core.agent_orchestrator import AgentOrchestrator
from models.openai_model import OpenAIModel
```

```python
# ✅ Step 4: Initialize the model and agent orchestrator
model = OpenAIModel(model="gpt-3.5-turbo")  # or gpt-3.5-turbo for cheaper runs
orchestrator = AgentOrchestrator(model=model)
```

```python
# ✅ Step 5: Choose a ticker and run analysis
ticker = "AAPL"
time_horizon = "both"  # options: "short_term", "long_term", "both"

results = orchestrator.analyze(ticker, time_horizon)
```

```python
# ✅ Step 6: Display Final Recommendation
from pprint import pprint

print("🧠 Final Recommendation:")
pprint(results["final_recommendation"])
```

```python
# ✅ Optional: Display Raw Agent Outputs

print("\n📊 Quant Analysis:")
pprint(results.get("quant", {}))

print("\n📚 Fundamentals Analysis:")
pprint(results.get("fundamentals", {}))

print("\n📰 News Summary:")
pprint(results.get("news", {}))
```
