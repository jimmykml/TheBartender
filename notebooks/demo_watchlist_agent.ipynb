# 🔔 WatchlistAgent Demo
This notebook demonstrates how to add, remove, and check stock price alerts using the WatchlistAgent.

```python
%load_ext autoreload
%autoreload 2

# ✅ Step 1: Setup import path
import os
import sys
from dotenv import load_dotenv

load_dotenv()

project_root = os.path.abspath(os.path.join(os.getcwd(), ".."))
os.chdir(project_root)  # Force working directory to project root

```

```python
# ✅ Step 2: Import WatchlistAgent
from agents.watchlist_agent import WatchlistAgent
agent = WatchlistAgent()
```

```python
# ✅ Step 3: Add stocks to watchlist
agent.add("AAPL", 200)
agent.add("TSLA", 750)

print("✅ Added AAPL and TSLA to watchlist.")
```

```python
# ✅ Step 4: Check current alerts
alerts = agent.check_alerts()

if alerts:
    for alert in alerts:
        print(f"🔔 {alert['message']}")
else:
    print("📭 No alerts triggered yet.")
```

```python
# ✅ Step 5: Remove stock from watchlist
agent.remove("TSLA")
print("🧹 Removed TSLA from watchlist.")
```
